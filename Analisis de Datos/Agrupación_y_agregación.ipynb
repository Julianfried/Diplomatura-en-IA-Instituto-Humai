## Parte A
Para esta parte vamos a trabajar sobre el mismo DataFrame que usamos en clase pero agrupando a nivel de comunidad autónoma (Cod_CCAA). En este caso vamos a traer también las descripciones y vamos a hacer un join para formar un único data set.

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('https://datasets-humai.s3.amazonaws.com/datasets/parodesprov.csv')
```

```python
df_descripciones = pd.read_csv('https://datasets-humai.s3.amazonaws.com/datasets/comunidades_descripcion.csv')
```

```python
df.head()
```

```python
df_descripciones
```

Unan ambos DataFrames utilizando la función pd.merge()

```python
merge = pd.merge(df, df_descripciones, left_on = 'Cod_CCAA', right_on = 'Código')
merge
```

Construir el DataFrame de comunidades autónomas, incluyendo la densidad, la proporción de paro y la cantidad de municipios. 

```python
merge['Cod_CCAA'].value_counts(dropna=False)
```

```python
len(df['Cod_CCAA'].unique())
```

```python
merge[merge.isnull().any(axis=1)]
```

```python
merge
```

```python
merge['Densidad'] = merge['PAD_1C02'] / merge['Shape__Area']
```

```python
merge['Proporcion_paro'] = merge['TotalParoRegistrado'] / merge['PAD_1C02']
```

```python
merge['Cantidad_muni'] = merge[]
```

```python
merge.groupby('Cod_CCAA').aggregate({
    'Densidad' : 'mean',
    'Proporcion_paro' : 'mean',
    'Codigo' : 'count'
    
}).sort_values('Densidad')
```

Habíamos visto que el promedio de la proporción de paro en los municipios con una densidad menor a la mediana es más bajo.. ¿Se cumple lo mismo para las provincias?

```python
mediana_densidad = merge['Densidad'].median()
```

```python
merge.query('Densidad >= @mediana_densidad')['Proporcion_paro'].mean()
```

```python
merge.query('Densidad < @mediana_densidad')['Proporcion_paro'].mean()
```

¿Cuál es la comunidad autónoma con mayor cantidad de municipios?

```python
cant_muni = merge.groupby('Cod_CCAA')['Codigo'].count().idxmax()

print(f'La C.A con mas municipios es la numero {cant_muni}')
```

```python
import numpy as np
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.5f' % x)
```

```python
df = pd.read_csv("https://cdn.buenosaires.gob.ar/datosabiertos/datasets/secretaria-de-desarrollo-urbano/terrenos-valor-oferta/Terrenos-en-venta-2020.csv", encoding = "latin1", sep=",")
```

```python
df.isnull().sum()
```

Hacer un primer análisis del dataset y analizar nulos.

```python
df['TRIMESTRE_'].unique()
```

```python
reemplazo = {
    'PRIMER' : 1,
    'SEGUNDO' : 2,
    'TERCER' : 3,
    'CUARTO' : 4
}
df['TRIMESTRE_'] = df['TRIMESTRE_'].replace(reemplazo)
```

```python
df['TRIMESTRE_'].unique()
```

### Ejercicio 1. 
¿Cuál es el precio medio del metro cuadrado en la Ciudad de Buenos Aires?

```python
precio_medio = round(df['DOLARM2'].mean()
, 2)
print(f'El precio metro cuadrado en la Ciudad de Buenos Aires es: {precio_medio}')
```

### Ejercicio 2. 
a) Dividir las propiedades por cuartiles de precios en dólares. 

```python
df['Cuartil'] = pd.qcut(df['PRECIOUSD'], 4, labels = ['Q1', 'Q2', 'Q3', 'Q4'])
```


b) ¿Cuáles son los barrios con más domicilios en venta dentro del cuartil más barato de CABA? 

Usar pivot_table

```python
tabla = pd.pivot_table(df, values='PROPIEDADS', index='BARRIO', columns='Cuartil', aggfunc='count')
```

```python
tabla
```

```python
barriosq1 = tabla['Q1'].sort_values(ascending = False)
print("Los 5 barrios con más domicilios en venta dentro del cuartil más barato son:")
print(barriosq1.head(5))
```

¿Y usando groupby?

```python
dfgroup = df.groupby(['BARRIO', 'Cuartil'])['DIRECCION'].count()

dfgroup.loc[:, 'Q1'].sort_values(ascending=False)
```

¿Hay otra forma más sencilla?

```python
df[df['Cuartil'] == 'Q1']['BARRIO'].value_counts().head()
```

### Ejercicio 3. 
¿Se puede calcular el tamaño de las propiedades? 



```python
df['Tamano'] = df['PRECIOUSD'] / df['DOLARM2']
df
```

¿Cuál sería el precio promedio de las propiedades para cada decil, según metros cuadrados? 

```python
deciles_m2 = pd.qcut(df['Tamano'], 10)
```

```python
precio_promedio_por_decil = df.groupby(deciles_m2)['DOLARM2'].mean().reset_index().sort_values(by="Tamano",ascending=True)
precio_promedio_por_decil
```

### Ejercicio 4.

¿Cuáles son los barrios más caros en términos de precios por metro cuadrado? ¿Y los más baratos? 


```python
precio_promedio_por_barrio = df.groupby('BARRIO')['DOLARM2'].mean().reset_index()
barrios_ordenados = precio_promedio_por_barrio.sort_values(by='DOLARM2',ascending=False)
barrios_ordenados.head()
```

```python
barrios_ordenados.tail()
```
