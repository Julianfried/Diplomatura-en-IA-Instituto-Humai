### d) ¿Cuál era la tasa de desocupación en el período seleccionado para las mujeres? ¿Era más alta o más baja que la de los hombres? (Extra: pensar el resultado usando una función)

Nota: la tasa se calcula como Desocupados / (Ocupados + Desocupados)

```python
def obtener_tasa_desocupacion(df):
  ocupados = df[df['estado'] == 1]
  desocupados = df[df['estado'] == 2]
  total_ocupados = ocupados['pondera'].sum()
  total_desocupados = desocupados['pondera'].sum()

  return round(total_desocupados/(total_desocupados + total_ocupados) * 100, 3)

print('La tasa de desocupacion total es:', obtener_tasa_desocupacion(eph_2))

tasa_varones = obtener_tasa_desocupacion(varones)
tasa_mujeres = obtener_tasa_desocupacion(mujeres)

print('La tasa de desocupaciones de varones es:', tasa_varones)
print('La tasa de desocupaciones de mujeres es:', tasa_mujeres)

if tasa_varones > tasa_mujeres:
  print('La tasa de varones es mayor')
else:
  print('La tasa de mujeres es mayor')
```

### e) Traer un dataframe que filtre a las personas ocupadas de más de 18 años y menores a 65 años.
- Obtener el promedio de ingresos
- Obtener el salario maximo ¿Cuál es su categoria ocupacional?
- Obtener el salario minimo ¿Cuál es su categoria ocupacional?

Nota: el promedio debe ser un promedio ponderado

$p_i = pondera \mid peso$

$v_i = valor$

Para facilitar el ejercicio se recomienda crea con el salario total por registro: 

\begin{align} salarioTot = v_i \cdot p_i \end{align}

Formula general:
\begin{align}
    Promedio Pond &= \frac{\sum_{i=1}^n v_i \cdot p_i}{\sum_{i=1}^n p_i } =  \frac{\sum_{i=1}^n salarioTot_i}{\sum_{i=1}^n p_i }\\
\end{align}

Ejemplo con 3 elementos:

\begin{align}
    Promedio Pond &= \frac{v_1 p_1 + v_2 p_2 + v_3 p_3}{p_1 + p_2 + p_3}.
\end{align}


```python
personas = eph_2.query('18 < edad < 65')

personas['salario_total'] = personas['ingreso'] * personas['pondera']

ocupaciones = {
  1: 'Patrón',
  2: 'Cuenta propia',
  3: 'Obrero o empleado',
  4: 'Trabajador familiar sin remuneración',
  9: 'Ns/Nr'
}
salario_total = personas[personas['salario_total'] > 0].copy()
promedio_ponderado = salario_total['salario_total'].sum()/salario_total['pondera'].sum()
print('El promedio ponderado es:', round(promedio_ponderado, 2))

salario_max = personas['ingreso'].max()
cat_max = salario_total.loc[salario_total['ingreso'].idxmax(), 'cat_ocup']
salario_min = personas['ingreso'].min()
cat_min = salario_total.loc[salario_total['ingreso'].idxmin(), 'cat_ocup']

print(f'El salario maximo es de {salario_max} pesos y su ocupacion es {ocupaciones[cat_max]}')
print(f'El salario minimo es de {salario_min} pesos y su ocupacion es {ocupaciones[cat_min]}')
```
