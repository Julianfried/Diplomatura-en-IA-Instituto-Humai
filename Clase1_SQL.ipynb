```python
%sql CREATE SCHEMA IF NOT EXISTS humai
```

```python

%sql CREATE TABLE IF NOT EXISTS humai.Customers (customer_id INT NOT NULL, customer_name VARCHAR(50) NOT NULL, fecha_inicio DATE NOT NULL, fecha_fin DATE, PRIMARY KEY (customer_id));
```

```python
%sql CREATE TABLE IF NOT EXISTS humai.Orders (order_id INT NOT NULL, customer_id INT NOT NULL, order_date DATE NOT NULL, order_price DECIMAL(8,2), PRIMARY KEY (order_id), FOREIGN KEY (customer_id) REFERENCES humai.Customers(customer_id));
```

```python
%sql CREATE TABLE IF NOT EXISTS humai.Shipments ( shipment_id INT NOT NULL, order_id INT NOT NULL, shipment_date DATE NOT NULL, shipment_city VARCHAR(50), PRIMARY KEY (shipment_id), FOREIGN KEY (order_id) REFERENCES humai.Orders(order_id));
```

```python
%sql SELECT count(*) AS "Ventas en 2022" FROM humai.Orders WHERE EXTRACT(YEAR FROM order_date) = 2022;
```

```python
%sql SELECT order_date AS "Fecha de primera venta" FROM humai.Orders ORDER BY order_date ASC LIMIT 1;
```

```python
%sql SELECT customer_id, MAX(order_price) as "Mayor Venta" FROM humai.Orders  GROUP BY customer_id;
```

```python
%sql SELECT shipment_city, count(shipment_id) as "n_shipments" FROM humai.Shipments WHERE shipment_city != 'Belgrano' GROUP BY shipment_city;
```

```python
%sql SELECT customer_name, CAST(date_part('day', NOW() - fecha_inicio) AS INTEGER) AS active_days FROM humai.Customers;
```

```python
%sql SELECT AVG(CAST(date_part('day', NOW() - fecha_inicio) AS INTEGER)) as "avg_active_days" FROM humai.Customers WHERE fecha_fin IS Null
```

```python
%sql SELECT COUNT(shipment_id) as n_shipments_mqd FROM humai.Shipments WHERE shipment_date >= '2022-01-01' AND shipment_date <= '2022-05-31' AND shipment_city = 'Mar del Plata';
```

```python
%sql SELECT customer_name, SUM(order_price) AS total_order_price FROM humai.Customers INNER JOIN humai.Orders ON humai.Customers.customer_id = humai.Orders.customer_id WHERE order_price > 20 GROUP BY humai.Customers.customer_id;
```
