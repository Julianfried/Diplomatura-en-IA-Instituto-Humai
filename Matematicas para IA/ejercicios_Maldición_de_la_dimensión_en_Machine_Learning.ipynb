## Volumen de vecindarios en distintas dimensiones
Usando Python, grafique cómo varía el tamaño del lado de un vecindario necesario para abarcar un determinado porcentaje 's' del volumen total de un hipercubo en función de la dimensión 'p'. Considere un 's' de 10 y 1 por ciento.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tamano_lado_vecindario(s,p):
  tamano_lado = s**(1/p)
  return tamano_lado

dimensiones = np.arange(1,21)
s1 = 0.01
s2 = 0.1

tamano_lado_s1 = [tamano_lado_vecindario(s1,p) for p in dimensiones]
tamano_lado_s2 = [tamano_lado_vecindario(s2,p) for p in dimensiones]

plt.plot(dimensiones, tamano_lado_s1, label='1%')
plt.plot(dimensiones, tamano_lado_s2, label='10%')
plt.xlabel('Dimensiones')
plt.ylabel('Tamano del lado del vecindario')
plt.legend()
plt.show()
```

## Incremento de esferas unitarias para cubrir el espacio A
Calcule y grafique cuántas esferas unitarias de dimensión p se necesitan para cubrir el espacio A para dimensiones entre 1 y 20.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

def num_esferas(p):
  return gamma(p/2+1) / np.pi**(p/2)

dimensiones = np.arange(1,21)
esferas = [num_esferas(p) for p in dimensiones]

plt.plot(dimensiones, esferas)
plt.xlabel("Dimensiones")
plt.ylabel("Número de esferas unitarias")
plt.title("Número de esferas unitarias para cubrir espacio A")
plt.show()
```

## Reducción de la dimensión usando PCA
Generar un conjunto de datos y realizar una reducción de dimensionalidad utilizando PCA. Represente gráficamente el resultado.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

rng = np.random.RandomState(1)
x1 = np.linspace(0,10,101)
x2 = 2 * x1 + rng.rand(101) * 5
x3 = x1 - x2 + rng.rand(101) * 2

datos = np.column_stack((x1,x2,x3))

pca = PCA(n_components = 2)
datos_transformados = pca.fit_transform(datos)

plt.scatter(datos_transformados[:, 0], datos_transformados[:,1])
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.title("Reducción de dimensionalidad usando PCA")
plt.show()
```
